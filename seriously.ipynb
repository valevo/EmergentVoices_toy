## creating relationships between comments

 - relationships can be on a scale (two extremes) and positively and negatively influence each other's chance
 - 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("answers.csv")

df = df[["id", "text", "language_question", "area_id", "choice_id", "subchoice_id", "user_name", "user_place", "time_created"]]
df = df[df.language_question.isin(["de", "en"])]
df["word_lists"] = df.text.apply(lambda s: ''.join([i if ord(i) < 128 else '' for i in s]))
df["word_lists"] = df["word_lists"].str.lower().str.split()
```
